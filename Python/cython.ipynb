__Пример1__:

Вычислим __сначала без Cython__:

```python
import numpy as np
from time import time

def max1(a, b):
    return a if a > b else b

def chebyshev1(x1, y1, x2, y2):
    return max1(abs(x1 - x2), abs(y1 - y2))

def benchmark():
    r=0
    np.random.seed(0)
    a1 = np.random.randint(100, size=3000) # одномерный массив на 3000
    b1 = np.random.randint(100, size=3000)  
    a2 = np.random.randint(100, size=3000)
    b2 = np.random.randint(100, size=3000)      
    
    for i in range(3000):
        for j in range(3000):
            r+=chebyshev1(a1[i], a2[j], b1[i], b2[j])

    return r

now = time()
print(benchmark())
print('Вычисления на чистом Python заняли:', time() - now)
```

```python
%%cython -a
import numpy as np
cimport cython

cdef int max(int a, int b):
    return a if a > b else b

cdef int chebyshev(int x1, int y1, int x2, int y2):
    return max(abs(x1 - x2), abs(y1 - y2))

def c_benchmark():
    cdef int r=0
    cdef int x1, y1, x2, y2
    cdef int[:] a1, b1, a2, b2
    
    np.random.seed(0)
    a1 = np.random.randint(100, size=3000) # одномерный массив на 3000
    b1 = np.random.randint(100, size=3000)  
    a2 = np.random.randint(100, size=3000)
    b2 = np.random.randint(100, size=3000)      

    for i in range(3000):
        for j in range(3000):
            r+=chebyshev(a1[i], a2[j], b1[i], b2[j])

    return r
```

```python
from time import time

now = time()
print(c_benchmark())
print('Вычисления с Cython заняли:', time() - now)
```

```python
import numpy as np
import random
from time import time
import pandas as pd

def Mood(a):
    y = np.zeros(20000, dtype=np.float64)
    r = 0
    for i in range(20000):
        if i % 2 == 0: 
            y[r] = random.random() + a
            r=+1
    return np.sum(y)**0.5


l = range(10000)

now = time()
x=pd.Series(map(Mood, l))
print(x)
print('Вычисления на чистом Python заняли:', time() - now)
```
